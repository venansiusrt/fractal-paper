```python
fs = 30
FIG = (30,20)
M = 1
```

```python
df1 = pd.read_excel("D://ITS//Mandelbrot Project//Violin Plot//Dataset K=3 and N=10000//New Sim Results+upd.xlsx", sheet_name = 0)
df2 = pd.read_excel("D://ITS//Mandelbrot Project//Violin Plot//Dataset K=3 and N=10000//New Sim Results+upd.xlsx", sheet_name = 1)
df3 = pd.read_excel("D://ITS//Mandelbrot Project//Violin Plot//Dataset K=3 and N=10000//New Sim Results+upd.xlsx", sheet_name = 2)
df4 = pd.read_excel("D://ITS//Mandelbrot Project//Violin Plot//Dataset K=3 and N=10000//New Sim Results+upd.xlsx", sheet_name = 3)

df1_julia = pd.read_excel("D://ITS//Mandelbrot Project//Violin Plot//Dataset K=3 and N=10000//New Sim Results+upd.xlsx", sheet_name = 4)
df2_julia = pd.read_excel("D://ITS//Mandelbrot Project//Violin Plot//Dataset K=3 and N=10000//New Sim Results+upd.xlsx", sheet_name = 5)
df3_julia = pd.read_excel("D://ITS//Mandelbrot Project//Violin Plot//Dataset K=3 and N=10000//New Sim Results+upd.xlsx", sheet_name = 6)
df4_julia = pd.read_excel("D://ITS//Mandelbrot Project//Violin Plot//Dataset K=3 and N=10000//New Sim Results+upd.xlsx", sheet_name = 7)

boundary1 = pd.read_excel("D://ITS//Mandelbrot Project//Violin Plot//Dataset K=3 and N=10000//New Sim Results+upd.xlsx", sheet_name = 8)
boundary2 = pd.read_excel("D://ITS//Mandelbrot Project//Violin Plot//Dataset K=3 and N=10000//New Sim Results+upd.xlsx", sheet_name = 9)
boundary3 = pd.read_excel("D://ITS//Mandelbrot Project//Violin Plot//Dataset K=3 and N=10000//New Sim Results+upd.xlsx", sheet_name = 10)
boundary4 = pd.read_excel("D://ITS//Mandelbrot Project//Violin Plot//Dataset K=3 and N=10000//New Sim Results+upd.xlsx", sheet_name = 11)

def broken_plot_BM_no_tuning(df, FIG, fs, M, LB, typ, bound):
    gs = gridspec.GridSpec(2, 1, height_ratios = [4, 1]) 
    fig = plt.figure(figsize = FIG)

    axU = plt.subplot(gs[0])
    axU.set_ylim(70,104)
    axU.spines['bottom'].set_visible(False)
    axU.yaxis.set_label_coords(-0.05, 0.25)
    axU.xaxis.tick_top()
    axU.tick_params(labeltop = False)
    axU.tick_params(axis = "x", labelsize = fs)
    axU.tick_params(axis = "y", labelsize = fs)
    sns.violinplot(x = 'method', y = 'test', data = df, ax = axU)
    plt.xlabel(' ')
    plt.ylabel('Testing Accuracy', fontsize = fs, fontweight = "bold")

    axB = plt.subplot(gs[1])
    axB.set_ylim(LB-1, LB+1)
    axB.spines['top'].set_visible(False)
    axB.xaxis.tick_bottom()
    axB.tick_params(axis = "x", labelsize = fs)
    axB.tick_params(axis = "y", labelsize = fs)
    sns.violinplot(x = 'method', y = 'test', data = df, ax = axB)
    plt.xlabel('Classifiers', fontsize = fs, fontweight = "bold", labelpad = 40)
    plt.ylabel(' ')

    d = .0025
    kwargs = dict(transform = axU.transAxes, color = 'k', clip_on = False)
    axU.plot((-d, +d), (-d, +d), **kwargs)
    axU.plot((1-d, 1+d), (-d, +d), **kwargs)

    kwargs.update(transform = axB.transAxes)
    axB.plot((-d, +d), (1-d, 1+d), **kwargs)
    axB.plot((1-d, 1+d), (1-d, 1+d), **kwargs)
    
    if bound == 1:
        fig.suptitle("Boundary of Mandelbrot Set, I = " + str(typ), fontsize = fs*M, fontweight = "bold")
    elif bound == 2:
        fig.suptitle("Mandelbrot Set, I = " + str(typ), fontsize = fs*M, fontweight = "bold")
    else:
        fig.suptitle("Julia Set, I = " + str(typ), fontsize = fs*M, fontweight = "bold")
    
    #plt.savefig("Violin-Boundary-Mandelbrot-Same-Model-with-Regular-"+str(typ)+".eps", format = 'eps')
```

```python
broken_plot_BM_no_tuning(boundary4, FIG, 45, M, 16, 4, 1)
plt.savefig('Violin-Boundary-Mandelbrot-Same-Model-with-Regular-4+upd.eps', format = 'eps')
```

```python
broken_plot_BM_no_tuning(df4, FIG, 45, M, 16, 4, 2)
plt.savefig('Violin-Mandelbrot-4+upd.eps', format = 'eps')
```
